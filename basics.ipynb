# Introduction to NetworkX 


- Network analysis is a broad, emerging field across many disciplines. 
This is a brief introduction to how to use the networkx library in python to do basic network analysis 




```python

import networkx as nx 
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np  


# NX - a built-in function from networkx that allows 
# you to create random networks 
# Barabasi-Albert graph is an example, it's a highly skewed graph with a few central nodes 

G = nx.barabasi_albert_graph(50, 2, seed=np.random.RandomState(10))

nx.draw_spring(G)
```

```python
erdos_renyi_graph = nx.erdos_renyi_graph(50, 0.6)  # nx.erdos_renyi_graph
nx.draw_spring(G)
```

```python
plt.hist([v for k,v in nx.degree(G)])


# In graph theory, betweenness centrality is a measure of centrality in a graph based on shortest paths
plt.hist(nx.centrality.betweenness_centrality(G).values())

```

```python
nx.diameter(G)
```

```python
nx.cluster.average_clustering(G)
```

```python
# values = G.nodes.data()
# print(f"Values:     {values}")

list(G.nodes(data=True))
```

# Importing from the Pandas library

- In order to create a graph object, you need to convert the data into what's called an "edgelist". The easiest way to do this is with a data frame which has at least two columns.

```python
nodes = list(range(1000))
df = pd.DataFrame({"from" : np.random.choice(nodes,100),
                    "to": np.random.choice(nodes,100)   
})
```

```python
df
```

```python
G = nx.from_pandas_edgelist(df, source="from", target="to")
```

```python
nx.draw(G)
```

```python
plt.hist(nx.centrality.betweenness_centrality(G).values())
```
